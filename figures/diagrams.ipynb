```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
execution_time_data = pd.read_csv('./execution-time.csv')

plt.rcParams.update({
    'font.family': 'Latin Modern Roman',
    'text.usetex': True,
    'font.size': 14
})
plt.figure(figsize=(9, 4), dpi=200)
plt.bar(execution_time_data['size'], execution_time_data['time'], edgecolor='black', color='white', width=100, zorder=1)

slope, intercept, r, p, std_err = stats.linregress(execution_time_data['size'], execution_time_data['time'])
line_x = [89, 1667]
line_y = [slope * x + intercept for x in line_x]

plt.plot(line_x, line_y, color='#ee2a44', label='Linear Regression Line', linestyle='--', zorder=2)
plt.xlabel('File Size (MB)')
plt.ylabel('Execution Time (s)')
plt.title(r'\textbf{Execution Time by File Size}')
plt.xticks(execution_time_data['size'])
plt.grid(axis='y', zorder=0)
plt.legend(frameon=False, columnspacing=1)

plt.savefig('execution-time.png', bbox_inches='tight')
plt.show()
```
